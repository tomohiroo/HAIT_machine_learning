```python
import pandas as pd
from sklearn import svm, metrics, cross_validation
```

```python
df = pd.read_csv('titanic_train.csv') #CSVデータの読み込み
```

```python
df.head() #先頭の5行を表示
```

```python
df.isnull().sum() #NaNの個数を表示
```

```python
sexmapping = {'male': 0, 'female': 1}　#文字をマッピング
df['Sex'] = df['Sex'].map(sexmapping)
embarked = {'C':0, 'Q':1, 'S':2}　　　　#文字をマッピング
df['Embarked'] = df['Embarked'].map(embarked)
```

```python
df = df.fillna(df.mean())　#平均値で欠損値に代入
```

```python
X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch','Embarked' ]]　#説明変数を抽出
y = df['Survived'] #目的変数を抽出
print(X) #なんとなく確認
print(y)
```

```python
#クロスバリデーションを使う
clf = svm.SVC()
scores = cross_validation.cross_val_score(clf, X, y, cv=15) #なんとなく15分割
scores.mean() #15分割の平均を出力
```
